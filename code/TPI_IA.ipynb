**Rellenado de NaN**

Vamos a rellenar los NaN utilizando la libreria Numpy. La logica que decidimos llevar es la de agrupar por sexo y edad, ya que son campos que consdieramos suficientemente representativos para tomar el promedio, cuyo valor es el que se va a reemplazar en el campo correspondiente a

```python
promedios_por_grupo = data.groupby(['sexo', 'edad'])['cigarrillos_por_dia'].transform('mean').fillna(0).astype(int)
data['cigarrillos_por_dia'].fillna(promedios_por_grupo, inplace=True)
```

```python
promedios_por_grupo = data.groupby(['sexo', 'edad'])['colesterol_total'].transform('mean').fillna(0).astype(int)
data['colesterol_total'].fillna(promedios_por_grupo, inplace=True)
```

```python
promedios_por_grupo = data.groupby(['sexo', 'edad'])['indice_masa_corporal'].transform('mean').fillna(0).astype(int)
data['indice_masa_corporal'].fillna(promedios_por_grupo, inplace=True)
```

```python
promedios_por_grupo = data.groupby(['sexo', 'edad'])['ritmo_cardiaco'].transform('mean').fillna(0).astype(int)
data['ritmo_cardiaco'].fillna(promedios_por_grupo, inplace=True)
```

```python
promedios_por_grupo = data.groupby(['sexo', 'edad'])['glucosa'].transform('mean').fillna(0).astype(int)
data['glucosa'].fillna(promedios_por_grupo, inplace=True)
```

```python
promedios_por_grupo = data.groupby(['sexo', 'edad'])['educacion'].transform('mean').fillna(0).astype(int)
data['educacion'].fillna(promedios_por_grupo, inplace=True)
```

```python
promedios_por_grupo = data.groupby(['sexo', 'edad'])['toma_medicamentos_pre_art'].transform(lambda x: 0 if (x == 0).sum() >= (x == 1).sum() else 1).fillna(0).astype(int)
data['toma_medicamentos_pre_art'].fillna(promedios_por_grupo, inplace=True)
data.head(10)
```

```python
data.to_csv('data_completa.csv', index=False)
```

### Estandarización de Atributos


Aqui tratamos la estadarización de atributos de la columna sexo, intercambiando M por 1 y F por 0, tambien tratamos la columna fumador dondes intercambiamos un YES por un 1 y un NO por un 0.


```python
data['sexo'] = data['sexo'].replace({'M': 1, 'F': 0})
data['fumador'] = data['fumador'].replace({'YES': 1, 'NO': 0})
```

```python
data.to_csv('DataFinal.csv', index=False)
```
